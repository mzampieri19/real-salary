Imported and opened the JSON object that was created in the preliminary analysis notebook. 

```python
import json

with open('selected_words_sector.json') as f:
    selected_words = json.load(f)

print(selected_words)
```

```python
# Print all the sectors
for sector in selected_words:
    print(sector)
    print(selected_words[sector])
    print("Number of words in sector: ", len(selected_words[sector]))
```

Turn the JSON object into a list of dictionaries, the sector names are the keys and the list of words are the values.

```python
list_of_dicts = [{'sector': sector, 'words': words} for sector, words in selected_words.items()]

# Print the list of dictionaries
for entry in list_of_dicts:
    print(entry)
```

Sort in alphabetical order.

```python
list_of_dicts.sort(key=lambda x: x['sector'])

for entry in list_of_dicts:
    print(entry)
```

```python
# Save the list of dictionaries to a json file
with open('selected_words_sector_list.json', 'w') as f:
    json.dump(list_of_dicts, f, indent=4)
```
