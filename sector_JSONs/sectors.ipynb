```python
# Get a list of all sector names
sector_names = [list(d.keys())[0] for d in dictionary_list]

# Put them in alphabetical order
sector_names.sort()
print(sector_names)
```

```python
def combine_sectors(dictionary_list, sector_pairs):
    combined_dict = {}
    
    for sector1, sector2 in sector_pairs:
        combined_name = f"{sector2}"
        
        # Find the matrices for the sectors
        matrix1 = None
        matrix2 = None
        
        for d in dictionary_list:
            if sector1 in d:
                matrix1 = np.array(d[sector1])
            if sector2 in d:
                matrix2 = np.array(d[sector2])

        print(f"Combining sectors: {sector1} and {sector2}")
        print(f"Matrix1: {matrix1}")
        print(f"Matrix2: {matrix2}")
        
        if matrix1 is not None and matrix2 is not None:
            # Average the matrices
            combined_matrix = (matrix1 + matrix2) / 2
            combined_dict[combined_name] = combined_matrix.tolist()
            print(f"Combined matrix for {combined_name}: {combined_matrix}")
        else:
            print(f"One or both matrices not found for sectors: {sector1}, {sector2}")

    
    return combined_dict

# List of sector pairs to combine
sector_pairs = [
    ("Aerospace_Defense", "Aerospace and Defense"),
    ("Appliances and Electronics", "Appliances"),
    ("Automotive Parts", "Automotives"),
    ("Flavor and Fragances", "Flavor_Fragances"),
    ("Food_Beverage", "Food_Beverages"),
    ("Metals and Mining", "Mining and Metals"),
    ("Oil_Gas", "Oill and Gas")
]

d_list = []
# Load the dictionary list from the JSON file
for sector_dict in dictionary_list:
    for sector in sector_dict.keys():
        for sector1, sector2 in sector_pairs:
            if sector1 == sector or sector2 == sector:
                d_list.append(sector_dict)
                break

# Sort the list of dictionaries by sector name
d_list.sort(key=lambda x: list(x.keys())[0])
print(d_list)


# Combine the sectors
#combined_sectors = combine_sectors(d_list, sector_pairs)

# Print the combined sectors
#print(combined_sectors)
```
